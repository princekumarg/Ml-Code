```python
import numpy as np
from scipy.cluster.vq import kmeans,whiten
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
pts=100
a=np.random.multivariate_normal([0,0],[[4,1],[1,4]],size=pts)
b=np.random.multivariate_normal([30,30],[[10,2],[2,1]],size=pts)
features=np.concatenate((a,b))
whitened=whiten(features)
codebook,distortion=kmeans(whitened,2)
plt.scatter(whitened[:,0],whitened[:,1])
plt.scatter(codebook[:,0],codebook[:,1],c='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")
fre=10
fre_samp=100
t=np.linspace(0,2,2*fre_samp,endpoint=False)
a=np.sin(fre*2*np.pi*t)
plt.plot(t,a)
plt.xlabel('Time(s)')
plt.ylabel('signal amplitude')
plt.show()
```

```python
from scipy import fftpack
A=fftpack.fft(a)
frequency=fftpack.fftfreq(len(a))*fre_samp
plt.stem(frequency,np.abs(A),use_line_collection=True)
plt.xlabel('Frequency in Hz')
plt.ylabel('Frequency spectrum Magnitude')
plt.show()
```

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
x=np.linspace(0,5,10)
y=np.cos(x**2/3+4)
plt.scatter(x,y,c='r')
plt.show()
```

### interpolate in scipy(use to find property of math function)

```python
#use of scipy interpolate
from scipy.interpolate import interp1d # that interp1(one)d
import matplotlib.pyplot as plt
fun1=interp1d(x,y,kind='linear')
fun2=interp1d(x,y,kind='cubic')
xnew=np.linspace(0,4,30)
plt.plot(x,y,'o',xnew,fun1(xnew),'-',xnew,fun2(xnew),'--')
plt.legend(['data','linear','cubic','nearest'],loc='best')
plt.show()
```

### linalg in scipy(linear algebra)

```python
import numpy as np
from scipy import linalg
a=np.array([[1,2,3],[2,-5,4],[5,4,-1]])
b=np.array([[-3],[13],[5]])
x=linalg.solve(a,b)
print(x)
print('\n checking result,must be zero')
print(a.dot(x)-b)
```

```python
import numpy as np
from scipy import linalg
A=np.array([[1,2,9],[3,4,8],[7,8,4]])
x=linalg.det(A)
print('determinant of \n{} \n is {}'.format(A,x))# we use .formate for float and ,formate for string
```
